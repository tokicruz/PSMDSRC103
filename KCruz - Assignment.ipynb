## Task 1: Type the expressions below, and try to explain what's happening in each case.

#### Set 1A

- `2 ** 16`
- `2 / 5, 2 / 5.0`
- `"spam" + "eggs"`
- `S = "ham"`
- `"eggs " + S`
- `S*5`
- `S[:0]`
- `"green %s and %s" % ("eggs", S)`

```python
# 2 ** 16
# Calculates 2 raised to the power of 16. The ** operator is used for exponentiation.
print (2 ** 16) 

# 2 / 5, 2 / 5.0
# Performs division, with the integer 2 as the numerator of both expressions.
# Integer 5 and floating-point number 5.0 as the denominators of each of the expressions, respectively.
print(2 / 5, 2 / 5.0)

# "spam" + "eggs"
# Concatenates or combines the two strings, "spam" and "egg".
print("spam" + "eggs")  

# S = "ham"
# Variable S gets assigned the value of the string "ham".
S = "ham"
print (S)

# "eggs " + S
# Concatenates the string "eggs " with the value of the variable S, which is the string "ham".
print("eggs " + S)

# S*5
# Repeats the value of the variable S, which is the string "ham", by 5 times.
print(S*5)

# S[:0]
# Slices the value of S, which is the string "ham". 
# Slicing starts and stops both at index 0, or the first character "h".
# Hence, returns with an empty string.
print(S[:0])

# "green %s and %s" % ("eggs", S)
# A format string, where %s acts as a placeholder for strings. 
# In their sequence, %s replaced by the corresponding values provided after the % operator.
# Values of "eggs" and of "ham", for the variable S, are inserted into the string.
print("green %s and %s" % ("eggs", S))
```

#### Set 1B

- `('x',) [0]`
- `('x', 'y') [1]`

```python
# ('x',) [0]
# Trailing comma signifies a single-element tuple, with the element 'x'.
# Note: Sans a comma, this is instead a string in parentheses. 
# Seeking to access the first element, at index 0, 'x' is returned.
print(('x',) [0])

# ('x', 'y') [1]
# A tuple with two elements, 'x' and 'y'.
# Seeking to access the second element, at index 1, 'y' is returned.
print(('x', 'y') [1])
```

#### Set 1C
- `L = [1,2,3] + [4,5,6] `
- `L, L[:], L[:0], L[-2], L[-2:]` 
- `([1,2,3] + [4,5,6]) [2:4]` 
- `[L[2], L[3]] `
- `L.reverse(); L` 
- `L.sort(); L `
- `L.index(4){'a':1, 'b':2} ['b']`

```python
# L = [1,2,3] + [4,5,6] 
# concatenates two lists, [1,2,3] and [4,5,6] 
# Combines the elements in the two lists into a single list, in sequence.
L = [1,2,3] + [4,5,6] 
print (L)

# L, L[:], L[:0], L[-2], L[-2:]
# Displays the list L in various slices
# L has all the elements in the list, that is [1, 2, 3, 4, 5, 6].
# L[:] uses a slice notation but nonetheless creates a copy of the entire list L... 
## (cont'n) ... absent start, stop, and step attributes in the slice notation.
# L[:0] is an empty list, slicing from start up to index 0.
# L[-2] displays the second-to-last element in the list L.
# L[-2:] slices from the second-to-last element to the last element in the list L.
print(L, L[:], L[:0], L[-2], L[-2:])

# ([1,2,3] + [4,5,6]) [2:4] 
# Initially concatenates the two lists [1,2,3] and [4,5,6].
# And further slices the resulting list to get elements at index 2 and 3.
print(([1,2,3] + [4,5,6]) [2:4])

# [L[2], L[3]] 
# Accesses the elements at index 2 and index 3, respectively, in the list L.
print([L[2], L[3]])

# L.reverse(); L 
# Reverses the list L in place, resulting to [6, 5, 4, 3, 2, 1].
L.reverse(); print(L)

# L.sort(); L 
# This sorts the (new) list L in place, reverting to [1, 2, 3, 4, 5, 6].
L.sort(); print(L)

# L.index(4){'a':1, 'b':2} ['b']
# An invalid syntax
# A dictionary cannot be placed immediately after a method call...
# (cont'n) ... without some form of operator or separation.
# But L.index(4) can be ran seprately.

# L.index(4)
# Presents the index of the first occurrence of the value 4 in the list L, which is index 3.
print(L.index(4))
```

#### Set 1D
- `D = {'x':1, 'y':2, 'z':3}`
- `D['w'] = 0`
- `D['x'] + D['w']`
- `D[(1,2,3)] = 4`
- `D.keys(), D.values(), D.has_key((1,2,3)) [[]], ["", [], {}, None]`

```python
# D = {'x':1, 'y':2, 'z':3}
# Creates dictionary D with keys 'x', 'y', and 'z'...
# (cont'n) ... and their corresponding values 1, 2, and 3.
D = {'x':1, 'y':2, 'z':3}
print (D)

# D['w'] = 0
# Adds a new key-value pair to dictionary D, that is 'w' with the value of 0.
# Results in {'x': 1, 'y': 2, 'z': 3, 'w': 0}
D['w'] = 0
print (D)

# D['x'] + D['w']
# Computes the sum of the values associated with keys 'x' and 'w'
print (D['x'] + D['w'])

# D[(1,2,3)] = 4
# Adds a new key in the form of the tuple (1, 2, 3), with the value 4
D[(1,2,3)] = 4
print (D)

# D.keys(), D.values(), D.has_key((1,2,3)) [[]], ["", [], {}, None]
# An invalid expression, but can be broken down for each piece to make sense. 

# D.keys()
# Presents all the keys in dictionary D.
print (D.keys())

# D.values()
# Presents all the values in dictionary D.
print (D.values())

# D.has_key((1, 2, 3))
# In Python 3, has_key() is obsolete. 
# Now the in operator can be used to check if key (1, 2, 3) does exist in dictionary D.
print((1, 2, 3) in D)

# [[]], ["", [], {}, None]
# A list containing different empty or "falsy" types.
# An empty string "".
# An empty list [].
# An empty dictionary {}.
# None represents no or null value.
print([[]], ["", [], {}, None])
```

## Task 2: Define a list named L that contains four strings or numbers. Experiment with the following boundary cases.

- Attempt to index out of bounds. (Ex. L[66])
- Attempt to slice out of bounds. (Ex. L[-9999:100])
- Try extracting a sequence in reverse. Make the lower bound higher than the upper bound (Ex. L[3:1]). Attempt to perform L[3:1] = ['?']. Explain the output.

```python
# L = [10, 20, 30, 40]
# Creates list L with numeric elements.
L = [10, 20, 30, 40]
print (L)

# Attempt to index out of bounds (example, L[66])
# IndexError raised because index 66 doesn't exist in the list.
try:
    print(L[66])
except IndexError:
    print("IndexError: Index 66 is out of range.")

# Attempt to slice out of bounds (example, L[-9999:100])
# Slicing from the 9999th-to-last element to the element at index 99, 
# (cont'n) ... from list L that consits of only 4 elements, does not raise an error.
# Only returns the elements within the range if they exist.
try:
    print(L[-9999:100])
except IndexError:
    print("IndexError: Slice indices are out of range.")

# Try extracting a sequence in reverse (example, L[3:1])
# Returns an empty list because the start index is greater than the stop index.
try:
    print(L[3:1])
except IndexError:
    print("IndexError: Slice indices are out of range.")

# Attempt to perform L[3:1] = ['?']
# Inserts '?' between index 3 and 1, effectively inserting it at index 3.
try:
    L[3:1] = ['?']
except IndexError:
    print("IndexError: Slice indices are out of range.")
else:
    print(L) 
```
